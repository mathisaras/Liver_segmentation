**Creating a group of 65 slices**

```python
in_path="D:/monai/dataset/dicomfiles/labels"
out_path="D:/monai/dataset/dicom_groups/labels"
```

```python
#!pip install glob2
#!pip install pytest-shutil
from glob import glob
import shutil
import os
```

```python
#patient_list=glob(in_path+'/*')
#print(patient_list[1])
#len(patient_list)
```

```python
for patient in glob(in_path + '/*'): #for every folder
     patient_name=os.path.basename(os.path.normpath(patient)) #=the foldername
     no_of_folders=int(len(glob(patient+'/*'))/64) #no.of.slices/minimum of minimum slices

     for i in range(no_of_folders):
      output_path_name=os.path.join(out_path,patient_name+'_'+str(i)) #create subfolders-(1-64) and the rest for each patient in patient list
      os.mkdir(output_path_name)
      for i, file in enumerate(glob(patient + '/*')):#moving in new folder
          if i == 64 + 1:
              break  #to have 65 slices in each folder
          shutil.move(file, output_path_name)
          
          
```

**Convert dicom groups into nifti**

```python
#!pip install dicom2nifti
import dicom2nifti
#!pip install nibabel
import nibabel as nib
#!pip install
import numpy as np
```

```python
in_path_images="D:/monai/dataset/dicom_groups/images/*"
in_path_labels="D:/monai/dataset/dicom_groups/labels/*"
out_path_images="D:/monai/dataset/nifti_files/images"
out_path_labels="D:/monai/dataset/nifti_files/labels"
list_images=glob(in_path_images)
list_labels=glob(in_path_labels)
```

```python
for patient in list_labels:
     patient_name=os.path.basename(os.path.normpath(patient)) #=the foldername
     dicom2nifti.dicom_series_to_nifti(patient, os.path.join(out_path_labels,patient_name+'.nii.gz'))
```
